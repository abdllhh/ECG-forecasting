```python

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load data
df_train = pd.read_csv('/content/ECG5000_TRAIN.txt', header=None, delim_whitespace=True)
df_test = pd.read_csv('/content/ECG5000_TEST.txt', header=None, delim_whitespace=True)
```

```python
# merging training and test data

df_merged = pd.concat([df_train, df_test], axis=0) #apparently df has no append attribute :)
print(df_merged.shape)
df_merged.head()
```

```python
# shuffling data

df_shuffled = df_merged.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled.head()
```

Overall dataset has 5 classes. One class is normal and other 4 classes are considered as anomalies

## Data Preprocessing


```python
CLASS_NORMAL = 1
class_names = ['Normal' , 'Anomaly1', 'Anomaly2', 'Anomaly3', 'Anomaly4']
new_columns = list(df_shuffled.columns)
new_columns[-1] = '140'
df_shuffled.columns = new_columns
df_shuffled.head()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
df_shuffled.Class.value_counts().plot(kind='bar')
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# converting to time-series using a function found on github ecg5000 getting-things-done-with-pytorch


def plot_time_series_class(data, class_name, ax, n_steps=10):
  time_series_df = pd.DataFrame(data)

  smooth_path = time_series_df.rolling(n_steps).mean()
  path_deviation = 2 * time_series_df.rolling(n_steps).std()

  under_line = (smooth_path - path_deviation)[0]
  over_line = (smooth_path + path_deviation)[0]

  ax.plot(smooth_path, linewidth=2)
  ax.fill_between(
    path_deviation.index,
    under_line,
    over_line,
    alpha=.125
  )
  ax.set_title(class_name)

classes = df_shuffled.Class.unique()

fig, axs = plt.subplots(
  nrows=len(classes) // 3 + 1,
  ncols=3,
  sharey=True,
  figsize=(14, 8)
)

for i, cls in enumerate(classes):
  ax = axs.flat[i]
  data = df_shuffled[df_shuffled.Class == cls] \
    .drop(labels='Class', axis=1) \
    .mean(axis=0) \
    .to_numpy()
  plot_time_series_class(data, class_names[i], ax)

fig.delaxes(axs.flat[-1])
fig.tight_layout();
```

## Data Preprocessing again to merge all anomalies as one, and normal as one

```python
normal_df = df_shuffled[df_shuffled.Class == 1] .drop(labels='Class', axis=1)
normal_df.shape

# Class_Normal = 1

```

```python
anomaly_df = df_shuffled[df_shuffled.Class != 1].drop(labels='Class', axis=1)
anomaly_df.shape
```

```python
# seperating data into training, test, validation using sklearns train_test_split

from sklearn.model_selection import train_test_split

RANDOM_SEED = 42

train_df, val_df = train_test_split(
  normal_df,
  test_size=0.15,
  random_state=RANDOM_SEED
)

val_df, test_df = train_test_split(
  val_df,
  test_size=0.33,
  random_state=RANDOM_SEED
)
```

```python
import pandas as pd
from darts import TimeSeries

# Assuming the time series data should have a regular frequency
# Create a datetime index with a uniform frequency
normal_df.index = pd.date_range(start='2023-01-01', periods=len(normal_df), freq='S')  # 'S' for second frequency

# Convert the DataFrame into a TimeSeries object
series = TimeSeries.from_dataframe(normal_df, fill_missing_dates=True)

# Optionally, split into training and validation sets
train, val = series.split_before(0.8)
```

```python
from darts.models import TCNModel

# Define a TCN model
model = TCNModel(
    input_chunk_length=30,  # Adjust based on sequence length
    output_chunk_length=10,  # Adjust based on sequence length
    n_epochs=50,
    dropout=0.2,
    dilation_base=2,
    weight_norm=True,
    kernel_size=3,
    num_filters=64,
    random_state=42
)

# Train the model on normal data
model.fit(train)
```

```python
# Get predictions on the validation set
predictions = model.predict(len(val), series=train)

# Calculate the absolute difference between the actual and predicted values
reconstruction_loss = np.abs(val.values() - predictions.values()).mean(axis=1)


# Set a threshold for anomaly detection
threshold = reconstruction_loss.mean() + 3 * reconstruction_loss.std()

# Identify anomalies
anomalies = reconstruction_loss > threshold

# Identify the indices where anomalies are detected
anomalies_indices = np.nonzero(anomalies)[0]

# Print the anomaly indices
print("Anomalies detected at:", anomalies_indices)
```

```python


# Plot the actual vs predicted series
plt.figure(figsize=(12, 6))
val.plot(label="actual")
predictions.plot(label="predicted")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.show()

# Plot the reconstruction loss
plt.figure(figsize=(12, 6))
plt.plot(reconstruction_loss, label="Reconstruction Loss")
plt.axhline(y=threshold, color='r', linestyle='--', label="Threshold")
plt.title("Reconstruction Loss with Anomaly Threshold")
plt.legend()
plt.show()
```
